# Descriptive statistics problems

### Exercise 1

We will use NumPy to obtain information to describe statistically.

- Generate an array of 100 elements following a normal distribution.
- Generate an array of 100 elements following a chi-square distribution with 3 degrees of freedom.
- Calculate the main metrics and statistical measures that best describe the two vectors.

````python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import statistics as stats
mean=0
std_dev = 1 
df=3
normal_array = np.random.normal(loc=mean, scale=std_dev, size=100)
chi_square_array = np.random.chisquare(df, size=100)
print(f"Normal vector: {normal_array}")
print(f"Chi-square vector: {chi_square_array}")
print(f"normal_array  mean: {stats.mean(normal_array)}")
print(f"chi_square_array  mean: {stats.mean(chi_square_array)}")
import numpy as np
from scipy import stats

# Example vectors
vector1 = np.random.normal(0, 1, 100)  # Vector 1 following a normal distribution
vector2 = np.random.chisquare(3, 100)  # Vector 2 following a chi-square distribution

# Function to calculate main statistical metrics
def calculate_metrics(vector):
    metrics = {
        "Mean": np.mean(vector),
        "Median": np.median(vector),
        "Variance": np.var(vector),
        "Standard Deviation": np.std(vector),
        "Minimum": np.min(vector),
        "Maximum": np.max(vector),
        "Range": np.ptp(vector),  # Range (max - min)
        "25th Percentile (Q1)": np.percentile(vector, 25),
        "75th Percentile (Q3)": np.percentile(vector, 75),
        "Interquartile Range (IQR)": stats.iqr(vector),
        "Skewness": stats.skew(vector),
        "Kurtosis": stats.kurtosis(vector)
    }
    return metrics

# Calculate metrics for both vectors
metrics_vector1 = calculate_metrics(vector1)
metrics_vector2 = calculate_metrics(vector2)

# Print metrics
print("Metrics for Vector 1:")
for key, value in metrics_vector1.items():
    print(f"{key}: {value:.4f}")

print("\nMetrics for Vector 2:")
for key, value in metrics_vector2.items():
    print(f"{key}: {value:.4f}")
````

### Exercise 2

Write a Python program to calculate the standard deviation of the following data:

```py
data = [4, 2, 5, 8, 6]
```

````python
import math
data = [4, 2, 5, 8, 6]

def sd_calc(data):
    n = len(data)
    if n <= 1:
        raise ValueError("At least two data points are required to calculate standard deviation.")

    mean = avg_calc(data)
    variance = sum((float(d) - mean) ** 2 for d in data) / (n - 1)
    return math.sqrt(variance)
    mean, sd = avg_calc(data), 0.0

    for d in data:
        sd += (float(d) - mean) ** 2
    sd = math.sqrt(sd / float(n - 1))

    return sd

def avg_calc(data):
    n, mean = len(data), 0.0

    if (n <= 1):
        return data[0]

    for d in data:
        mean = mean + float(d)

    mean = mean / float(n)
    return mean

print(f"Sample Data: {data}")
print(f"Standard Deviation: {sd_calc(data)}")
````
